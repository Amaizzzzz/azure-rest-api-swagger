```python
import os
import json


def get_openapi_json_file_paths(root_path):
    openapi_json_files = []
    filtered_files = []

    for dirpath, _, filenames in os.walk(root_path):
        for filename in filenames:
            # TODO: handle data-plane APIs later
            if filename.endswith(".json") and dirpath.find("stable") != -1 and dirpath.find("data-plane") == -1:
                file_path = os.path.join(dirpath, filename)
                parts = file_path.split("/")
                
                try:
                    with open(file_path, "r") as file:
                        json_data = json.load(file)
                        if "swagger" in json_data and json_data["swagger"].startswith("2.0"):
                            openapi_json_files.append(file_path)
                except json.JSONDecodeError:
                    pass
                except UnicodeDecodeError:
                    pass

    return openapi_json_files

# Replace this with the path to the root of the azure-rest-api-specs submodule in your local repo
azure_specs_path = "azure-rest-api-specs/specification/"

openapi_files = get_openapi_json_file_paths(azure_specs_path)

# Print the list of OpenAPI JSON file paths
for file_path in openapi_files:
    print(file_path)

```

```python

# Lambda function to slice out the version after /stable/
slice_version = lambda file_path: file_path.split("/stable/")[1].split("/")[0]
slice_version("./azure-rest-api-specs/specification/policyinsights/resource-manager/Microsoft.PolicyInsights/stable/2021-10-01/remediations.json")

```

```python
# use a dict: set the split("/stable/")[0] as key and update the version date to the most up to date
# Assuming you already have the filtered_files list obtained from the get_openapi_json_file_paths function

def is_version_newer(new_version, existing_version):
    new_version_parts = [int(part) for part in new_version.split("-")]
    existing_version_parts = [int(part) for part in existing_version.split("-")]

    if new_version_parts > existing_version_parts:
        return True
    return False

def find_latest_stable_versions(filtered_files):
    version_dict = {}

    for file_path in filtered_files:
        resource_version = file_path.split("/stable/")[0]
        date_part = file_path.split("/stable/")[1].split("/")[0]
        if resource_version not in version_dict or date_part > version_dict[resource_version]:
            version_dict[resource_version] = date_part
    
    return version_dict
latest_apis = find_latest_stable_versions(openapi_files)
latest_apis
```

```python
def is_of_latest_api(file_path):
    resource_version = file_path.split("/stable/")[0]
    date_part = file_path.split("/stable/")[1].split("/")[0]
    if date_part == latest_apis[resource_version]:
        return True
    return False

openapi_files_of_latest_api = list(filter(is_of_latest_api, openapi_files))
openapi_files_of_latest_api
```

```python
def find_friendly_name(file_path):
    resource_provider_part = file_path.split("/resource-manager/")[1].split("/")[0]
    file_name_without_json_extension = file_path.split("/")[-1].split(".json")[0]
    if resource_provider_part in friendly_names:
        # determine if friendly_names[resource_provider_part] is a string
        if isinstance(friendly_names[resource_provider_part], str):
            return f"{friendly_names[resource_provider_part]} ({file_name_without_json_extension})"
        elif resource_provider_part == "Microsoft.Compute":
            if file_name_without_json_extension == "virtualMachineScaleSet":
                return f"Virtual Machine Scale Sets ({file_name_without_json_extension})"
            else:
                return f"Virtual Machines ({file_name_without_json_extension})"
        else:
            # TODO: find the proper friendly name mapping
            return f"{resource_provider_part} ({file_name_without_json_extension})"
    else:
        return f"{resource_provider_part} ({file_name_without_json_extension})"

def convert_to_entry(file_path):
    # convert to a dictionary object where url is a github path like this:
    # https://raw.githubusercontent.com/Azure/azure-rest-api-specs/main/specification/compute/resource-manager/Microsoft.Compute/ComputeRP/stable/2023-03-01/virtualMachine.json
    # and name is the friendly name
    file_path_after_specification = file_path.split("specification/")[1]
    api_version = slice_version(file_path_after_specification)
    return {
        "url": f"https://raw.githubusercontent.com/Azure/azure-rest-api-specs/main/specification/{file_path_after_specification}",
        "api_version": api_version,
        "path_without_api_version": f"{file_path.split('/stable/')[0]}/{'/'.join(file_path.split('/stable/')[1].split('/')[1:])}",
        "name": find_friendly_name(file_path)
    }

def group_entries(entries_sorted_by_path_without_api_version):
    from itertools import groupby
    from operator import itemgetter
    return [{"label": find_friendly_name(key), "path_without_api_version": key, "api_version_options": list(sorted(group, key=lambda x: x["api_version"], reverse=True))} for key, group in groupby(entries_sorted_by_path_without_api_version, itemgetter("path_without_api_version"))]


sorted_entries = sorted(list(map(convert_to_entry, openapi_files)), key=lambda x: x["path_without_api_version"])

final_result = group_entries(sorted_entries)
final_result
```
