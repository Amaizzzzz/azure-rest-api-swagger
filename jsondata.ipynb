```python
import os
import json

def get_openapi_json_file_paths(root_path):
    openapi_json_files = []

    for dirpath, _, filenames in os.walk(root_path):
        for filename in filenames:
            if filename.endswith(".json") and dirpath.find("stable") != -1:
                file_path = os.path.join(dirpath, filename)
                try:
                    with open(file_path, "r") as file:
                        json_data = json.load(file)
                        if "swagger" in json_data and json_data["swagger"].startswith("2.0"):
                            openapi_json_files.append(file_path)
                except json.JSONDecodeError:
                    pass
                except UnicodeDecodeError:
                    pass

    return openapi_json_files

# Replace this with the path to the root of the azure-rest-api-specs submodule in your local repo
azure_specs_path = "./azure-rest-api-specs/specification/"

openapi_files = get_openapi_json_file_paths(azure_specs_path)

# Print the list of OpenAPI JSON file paths
for file_path in openapi_files:
    print(file_path)
```
